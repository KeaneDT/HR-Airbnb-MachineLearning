## 1. HR Analytics (Classification)

The HR Analytics dataset is a Logistic Regression problem that looks at employee data in order to predict if the employee will be promoted or not, through the use of the variable is_promoted where 1 is yes and 0 is no. The variables present in this cleaned dataset are:

**• department: 1 = Sales & Marketing, 2 = Operations, 3 = Technology, 4 = Analytics, 5 = R&D, 6 = Procurement, 7 = Finance, 8 = HR, 9 = Legal**

**• region: Region the Employee resides in, ranges from 1 to 34**

**• gender: Gender of the Employee, 0 = Female, 1 = Male**

**• recruitment_channel: Channel from which Employee was recruited, 1 = Sourcing , 2 = Referred, 3 = Other**

**• no_of_trainings: # of trainings, min-value = 1, max-value = 10**

**• age: Age of Employees, ranges from 20 to 60 years**

**• previous_year_rating: The rating the employee got in the previous year, ranges from 0-5, where 0 is for employees who joined within the year**

**• length_of_service: Length of service of the employee, ranges from 1 to 34 years**

**• kPIs_met_>80%: Whether the Employee has met more that 80% of their KPIs, 1 = Yes, 0 = No**

**• awards_won?: Whether the Employee has won any awards, 1 = Yes, 0 = No**

**• avg_training_score: Average score of the employee in current training evaluations, ranges from 41 to 99**

**• is_promoted: The target for the model, whether the Employee is promoted, 1 = Yes, 0 = No**

**• education_below_secondary: Whether the Employee has an education of Below Secondary, 1 = Yes, 0 = No**

**• education_bachelors: Whether the Employee has an education of Bachelor's, 1 = Yes, 0 = No, if yes also has below secondary**

**• education_masters_&_above: Whether the Employee has an education of Masters & Above, 1 = Yes, 0 = No, if yes also has bachelor's and below secondary**

## 1.2 Build the Model(s)

As the target for the HR Dataset is a categorical one, the values can only be either 0 or 1 and as such, the method to test the validity of the classifier models being built will be predictive accuracy.

Some additional factors to consider when assessing the models are the speed, robustness, scalability and interpretability.

### 1.2.1 Define Model Inputs (X) & Output (y)

Defining the Inputs for the models is very important as it will be used throughout the entire process. Here I defined X1 to be the variable containing all the variables except for the target, is_promoted. For y1, I defined it to be only the variable, is_promoted.

```python
X1 = df1.drop('is_promoted',axis=1)
y1 = df1["is_promoted"]
```

```python
X1.head()
```

```python
# Split both Inputs (X1) and Ouput (y1) into training set (70%) and testing set (30%)
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.3, random_state=2)
```

### 1.2.2 Model 1A Logistic Regression

The Logistic Regression Model plots the probility of each employee being promoted on a logistic Curve. This is done through the use of the formula, y=1/1+e^-(B_{0}+B_{1}x_{1}+...+B_{n}x_{n}), where x is the input and B is the coefficient. A decision boundary of 0.5 will also be used for the logistic curve. What this means is that if the probability of an employee is above 0.5, the predicted is_promoted will be a 1, vice versa.

#### 1.2.2.1 Creating the Model/Fitting to Training Set

```python
# Create Logistic Regression Model, lg
lg = LogisticRegression(solver='lbfgs', max_iter=10000)

# Fit the model to the training set
lg.fit(X1_train,y1_train)
```

#### 1.2.2.2 Calculate Fitted Values

```python
y1_fitted = lg.predict(X1_train) # returns 0 or 1
y1_fitted_prob = lg.predict_proba(X1_train)[:,1] # returns probabilities of 1, employee is promoted
```

```python
print(y1_fitted)
print(y1_fitted_prob)
```

#### 1.2.2.3 Accuracy Scores

```python
print(lg.score(X1_train, y1_train), '(Train Accuracy)')
print(lg.score(X1_test, y1_test), '(Test Accuracy)')
```

Here we can see that there is a relatively high bias for both the Training and Testing accuracy at around 0.72. There is also a low variance as both the Training and Testing accuracy are close to each other, with a difference of 0.00395. Hence this model is a decent one and is suitable to be used. Some Feature Selection based on P-Values and utilising GridSearchCV to find the optimal parameters can be used to increase the accuracy of this model.

### 1.2.3 Model 1B Decision Tree

The Decision Tree model

```python
decision_tree = tree.DecisionTreeClassifier(max_depth = 2)
decision_tree.fit(X1_train, y1_train)
```

```python
plt.figure(figsize=(25,10))
tree.plot_tree(decision_tree,
              feature_names = list(df1.drop(['is_promoted'], axis=1)),
                class_names = ['Not Promoted', 'Promoted'],
                rounded = True,
                filled= True)
plt.savefig('df1tree.png',format='png')
plt.show()  
```

```python
#training acc
train_acc = decision_tree.score(X1_train, y1_train)
test_acc = decision_tree.score(X1_test, y1_test)
print(train_acc, '(Train Accuracy)')
print(test_acc, '(Test Accuracy)')
```

### 1.2.4 Model 1C Artificial Neural Network

```python
# Create the ann model
mlp = MLPClassifier(activation ='logistic', hidden_layer_sizes=(10,), max_iter= 2000, solver = 'sgd', random_state=2)
# Fit the model to the training set
mlp.fit(X1_train,y1_train)

# Calculated the fitted values for training set
y1_fitted = mlp.predict(X1_train) # returns 0 or 1
y1_fitted_prob = mlp.predict_proba(X1_train)[:,0] # returns probabilities
```

### 1.3.5 Model 1D Random Forest - Tuning

#### 1.3.5.1 K-Fold Validation

```python
results  = cross_validate(rf, X1, y1, scoring='accuracy', cv=5, return_train_score = True)
print('train_score: ', results['train_score'])
print('test_score: ', results['test_score'])
```

```python
print('cross val training accuracy is:', sum(results['train_score'])/len(results['train_score']))
print('cross val testing accuracy is:', sum(results['test_score'])/len(results['test_score']))

# Variance still high which means the model is at fault
```

#### 1.3.5.2 Creating Model & Parameter Grid

```python
rf = RandomForestClassifier(max_features='auto', random_state=2, n_jobs=-1)

param_grid = { "criterion" : ["gini", "entropy"], 
              "max_depth": [2, 4, 6, 8], 
              "min_samples_leaf" : [1, 5, 10], 
              "n_estimators": [10, 20, 50, 100]}
```

#### 1.3.5.3 Using GridSearchCV to find the best parameters

```python
gs = GridSearchCV(estimator=rf, param_grid=param_grid, scoring='accuracy', cv=5, n_jobs=-1)

gs = gs.fit(X1, y1)

print(gs.best_score_)
print(gs.best_params_)
```

#### 1.3.5.4 Creating New Model using Best Parameters

```python
rf_final = RandomForestClassifier(criterion='entropy', 
                             n_estimators=10,
                             max_depth=8,
                             min_samples_leaf=1,
                             max_features='auto',                             
                             random_state=2,
                             n_jobs=-1)

rf_final.fit(X1_train,y1_train)

print('***Final Model***')
print('training acc: ', rf_final.score(X1_train,y1_train))
print('testing acc: ', rf_final.score(X1_test,y1_test))
results  = cross_validate(rf_final, X1, y1, scoring='accuracy', cv=5, return_train_score = True)
print('cross val training accuracy is:', sum(results['train_score'])/len(results['train_score']))
print('cross val testing accuracy is:', sum(results['test_score'])/len(results['test_score']))

# Better Variance
```
