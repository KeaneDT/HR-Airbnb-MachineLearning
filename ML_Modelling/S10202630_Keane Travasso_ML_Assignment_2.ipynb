## 1. HR Analytics (Classification)

The HR Analytics dataset is a Classification problem that looks at employee data in order to predict if the employee will be promoted or not, through the use of the variable is_promoted where 1 is yes and 0 is no. The variables present in this cleaned dataset are:

**• department: 1 = Sales & Marketing, 2 = Operations, 3 = Technology, 4 = Analytics, 5 = R&D, 6 = Procurement, 7 = Finance, 8 = HR, 9 = Legal**

**• region: Region the Employee resides in, ranges from 1 to 34**

**• gender: Gender of the Employee, 0 = Female, 1 = Male**

**• recruitment_channel: Channel from which Employee was recruited, 1 = Sourcing , 2 = Referred, 3 = Other**

**• no_of_trainings: # of trainings, min-value = 1, max-value = 10**

**• age: Age of Employees, ranges from 20 to 60 years**

**• previous_year_rating: The rating the employee got in the previous year, ranges from 0-5, where 0 is for employees who joined within the year**

**• length_of_service: Length of service of the employee, ranges from 1 to 34 years**

**• kPIs_met_>80%: Whether the Employee has met more that 80% of their KPIs, 1 = Yes, 0 = No**

**• awards_won?: Whether the Employee has won any awards, 1 = Yes, 0 = No**

**• avg_training_score: Average score of the employee in current training evaluations, ranges from 41 to 99**

**• is_promoted: The target for the model, whether the Employee is promoted, 1 = Yes, 0 = No**

**• education_below_secondary: Whether the Employee has an education of Below Secondary, 1 = Yes, 0 = No**

**• education_bachelors: Whether the Employee has an education of Bachelor's, 1 = Yes, 0 = No, if yes also has below secondary**

**• education_masters_&_above: Whether the Employee has an education of Masters & Above, 1 = Yes, 0 = No, if yes also has bachelor's and below secondary**

### Model 1A Logistic Regression

The Logistic Regression Model plots the probility of each employee being promoted on a logistic Curve. This is done through the use of the formula, y=1/1+e^-(B_{0}+B_{1}x_{1}+...+B_{n}x_{n}), where x is the input and B is the coefficient. A decision boundary of 0.5 will also be used for the logistic curve. What this means is that if the probability of an employee is above 0.5, the predicted is_promoted will be a 1, vice versa.

#### Define Model Inputs (X) & Output (y)

Defining the Inputs for the models is very important as it will be used as a baseline throughout the entire process. Here I defined X1 to be the variable containing all the variables except for the target, is_promoted. For y1, I defined it to be only the variable, is_promoted.

```python
X1 = df1.drop('is_promoted',axis=1)
y1 = df1["is_promoted"]
```

```python
X1.head()
```

```python
# Split both Inputs (X1) and Ouput (y1) into training set (70%) and testing set (30%)
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.3, random_state=2)
```

```python
# Check that the Data was split correctly
for sets in [y1_train, y1_test]:
    print(round(len(sets)/len(y1),2))
```

#### Baseline Logistic Regression Model

A Baseline Logistic Regression Model is created with the only hyperparameter being 10000 max iterations. The Training and Testing scores of the baseline model will be shown.

```python
# Create Logistic Regression Model, lg
lg = LogisticRegression(max_iter=10000)

# Fit the model to the training set
lg.fit(X1_train,y1_train)

printmd("**Accuracy Scores**")
print(lg.score(X1_train, y1_train), '(Train Accuracy)')
print(lg.score(X1_test, y1_test), '(Test Accuracy)')
```

#### K-Fold Cross Validation

K-Fold Cross Validation is a resampling technique that divides the total data into K folds/subsets at random to evaluate machine learning models. These folds/subsets are then trained and evaluated a total of K times, with each fold/subset serving as the test set only once. In comparison to the Simple Split method, which only allows for one split, this allows for a more accurate representation of the data to see what can be expected for the baseline performance.

```python
results  = cross_validate(lg, X1, y1, scoring='accuracy', cv=5, return_train_score = True)
printmd("**K-Fold Accuracy Scores (5 Fold)**")
printmd("**Train**")
print(results['train_score'], '(Train Score)')
print('The Average Train Score is:',np.mean(results['train_score']))

print(" ")

printmd("**Test**")
print(results['test_score'], '(Test Score)')
print('The Average Test Score is:',np.mean(results['test_score']))
```

Here we can see that there is a relatively low bias for both the Training and Testing accuracy as both of them are high values at around 0.72. There is also a low variance as both the Training and Testing accuracy are close to each other, with a difference of 0.00395. Hence this model is a decent one and is suitable to be used. Some Feature Selection based on P-Values and utilising GridSearchCV to find the optimal parameters can be used to increase the accuracy of this model.

#### Tuning of Hyperparameters (GridSearchCV)

Hyperparameters are model-external configurations whose values cannot be estimated from data and must instead be inputted by the user or by GridSearchCV.
```python
gs_lg_df = pd.DataFrame(gs_lg.cv_results_)
gs_lg_df.dropna().sort_values(by=["mean_test_score"],ascending=False)[['param_penalty','param_C','param_solver','param_max_iter','mean_test_score']].head(5)
```

### Model 1B Decision Tree

The Decision Tree Classifier Model is used to predict the outcome through the use of decision nodes and impurity. This is accomplished by first determining all possible binary split points as well as the best split for each of the input variables. The model then chooses the best split point with the lowest impurity reduction, as determined by the Gini Impurity Index. Based on the split, the data is then divided into subgroups. This cycle continues until all input variables are split or the stopping condition is met.

### Model 1C Artificial Neural Network

To predict an outcome, the MLP Classifier Model employs neurons in a multilayer structure which receive various inputs and weights using an Activation Function. The multilayer structure enables the algorithm to iterate learning by using the feedforward operation, in which data is passed through the various layers of neurons to compute the error. The result is then backpropagated, and all connection weights are updated, thereby stimulating learning. The process is repeated until a stop criterion is met.

```python
# Split both Inputs (X2) and Ouput (y2) into training set (70%) and testing set (30%)
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.3, random_state=2)
```

#### Baseline Random Forest Regressor Model

```python
rfr = RandomForestRegressor(max_features='auto', random_state=2, n_jobs=-1)
rfr.fit(X2_train, y2_train)
```

```python
train_rmse = sqrt(mean_squared_error(rfr.predict(X2_train), y2_train))
train_R2 = rfr.score(X2_train, y2_train)
printmd("**Train Scores**")
print(train_rmse, '(Train RMSE)') # Lower = Better
print(train_R2, '(Train R2)') # 0-1, Higher = Better
```

```python
test_rmse = sqrt(mean_squared_error(rfr.predict(X2_test), y2_test))
test_R2 = rfr.score(X2_test, y2_test)
printmd("**Test Scores**")
print(test_rmse, '(Test RMSE)') # Lower = Better
print(test_R2, '(Test R2)') # 0-1, Higher = Better
```

#### K-Fold Cross Validation

K-Fold Cross Validation is a resampling technique that divides the total data into K folds/subsets at random to evaluate machine learning models. These folds/subsets are then trained and evaluated a total of K times, with each fold/subset serving as the test set only once. In comparison to the Simple Split method, which only allows for one split, this allows for a more accurate representation of the data to see what can be expected for the baseline performance.

```python
results_rmse  = cross_validate(rfr, X2, y2, scoring='neg_root_mean_squared_error', cv=5, return_train_score = True)
results_r2  = cross_validate(rfr, X2, y2, scoring='r2', cv=5, return_train_score = True)

printmd("**K-Fold Accuracy Scores (5 Fold)**")
printmd("**Train**")
print(results_rmse['train_score'], '(RMSE Train Score)')
print(results_r2['train_score'], '(R2 Train Score)')

print('The Average RMSE Train Score is:',np.mean(-results_rmse['train_score']))
print('The Average R2 Train Score is:',np.mean(results_r2['train_score']))

print(" ")

printmd("**Test**")
print(results_rmse['test_score'], '(RMSE Test Score)')
print(results_r2['test_score'], '(R2 Test Score)')

print('The Average RMSE Test Score is:',np.mean(-results_rmse['test_score']))
print('The Average R2 Test Score is:',np.mean(results_r2['test_score']))
```
